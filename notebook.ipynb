```python
# read the excel sheets as data frams 
Balance_sheet= pd.read_excel('data/Balance_Sheet.xlsx')
Income_Statement = pd.read_excel('data/Income_Statement.xlsx')
```

```python
# merge the data frams of Balance sheets and Income Statements for calculating the financial ratios
Whole_DF=pd.merge(Balance_sheet,Income_Statement, on=['Year','company'])
print(Whole_DF.columns)
```

```python
# calculat the ratios
df_ratios = pd.DataFrame()
df_ratios[['Year','compt_type_x','company']] = Whole_DF[['Year','comp_type_x','company']]
df_ratios['leverage_ratio'] = Whole_DF['Total Liab'] / Whole_DF['Total Stockholder Equity']
df_ratios['profitability_ratio'] = (Whole_DF['Total Revenue'] - Whole_DF['Cost Of Goods Sold']) / Whole_DF['Total Revenue']
df_ratios.head(2)
```
